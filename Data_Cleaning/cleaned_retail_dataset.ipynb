## **Data cleaning**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
df = pd.read_excel("enterprise_retail_dataset_102500.xlsx")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Customer_Age'] = df['Customer_Age'].fillna(df['Customer_Age'].median())

df['Profit'] = df['Profit'].fillna(df['Profit'].median())

df['Rating'] = df['Rating'].fillna(df['Rating'].median())
```

```python
df['Payment_Method'] = df['Payment_Method'].fillna(
    df['Payment_Method'].mode()[0]
)
```

```python
df.duplicated().sum()
```

```python
df['Category'].unique()
```

```python
df['Category'] = df['Category'].replace({
    'Electronic': 'Electronics',
    'clothing': 'Clothing',
    'home and Lifestyle': 'Home & Lifestyle'
})
```

```python
df['Payment_Method'].unique()
```

```python
df[(df['Customer_Age'] < 18) | (df['Customer_Age'] > 80)]
```

```python
median_age = df['Customer_Age'].median()

df.loc[
    (df['Customer_Age'] < 18) |
    (df['Customer_Age'] > 80),
    'Customer_Age'
] = median_age
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'])
```

```python
df[['Order_Date','Delivery_Date']].dtypes
```

```python
df.to_excel("cleaned_enterprise_retail_dataset.xlsx", index=False)
```
